```python
df['PatientId'] = df['PatientId']. astype (str)
df['AppointmentID'] = df['AppointmentID']. astype (str)
df['Scholarship'] = df['Scholarship']. astype (bool)
df['Hipertension'] = df['Hipertension']. astype (bool)
df['Alcoholism'] = df['Alcoholism']. astype (bool)
df['Handcap'] = df['Handcap']. astype (bool)
df['SMS_received'] = df['SMS_received']. astype (bool)
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
df['Gender'] = df['Gender']. astype(object)
df['Diabetes'] = df['Diabetes']. astype(bool)
```

```python
df['WaitingDays'] = (df['AppointmentDay'] - df['ScheduledDay']).dt.days
```

```python
df['AppointmentDay'] = df['AppointmentDay'].dt.day_name()
```

```python
df = df[df['Age']>=0]
```

```python
age_bin = [0,10, 20,40,60,80,100,120]
age_labels = ['Kids', 'Teens', 'Adults', 'Mid-age', 'Old', 'Older', 'Super-old']
df['AgeGroup'] = pd.cut(df['Age'], bins = age_bin, labels= age_labels,include_lowest=True)
```

```python
df.head()
```

```python
df.groupby('AgeGroup')['Age'].count().sort_index()
```

```python
df['AgeGroup'].value_counts().sort_index().plot(kind='bar')
```

```python
plt.figure(figsize=(10,6))                      # Size of the plot
plt.bar(df['AgeGroup'].value_counts().index,    # x=labels
        df['AgeGroup'].value_counts().values,           # y=counts
        color='skyblue', edgecolor='black')    # colors and borders
plt.title("Number of Patients No-Showed by Age Group")   # chart title
plt.xlabel("Age Group")                        # x-axis label
plt.ylabel("Count of Patients")                # y-axis label
plt.xticks(rotation=45)                       # rotate labels 
plt.grid(axis='y', linestyle='--') 
plt.show()
```
